```python
X = df.drop(columns=['Type'])
y = df['Type']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

cv = StratifiedKFold(n_splits=3,shuffle=True,random_state=42)

models = {
    'Logistic Regression':LogisticRegression(max_iter=1000, multi_class='ovr' ),
    'Decision Tree':DecisionTreeClassifier(),
    'Random Forest':RandomForestClassifier(),
    'SVC':SVC(),
    'KNN':KNeighborsClassifier()
}
```

```python
results = {}
for name, model in models.items():
    cv_results = cross_val_score(model,X_train,y_train,cv=cv,scoring='accuracy')
    results[name] = cv_results
    print(f"{name}: {cv_results} | Mean: {np.mean(cv_results):.7f}")
```

```python
# Random Forest has the best accuracy
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import GridSearchCV

rf = RandomForestClassifier()
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 6, 12],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': ['sqrt', 'log2'],
    'bootstrap': [True]
}

from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
rf = RandomForestClassifier(random_state=42, n_jobs=-1)

gs = GridSearchCV(rf,
                  param_grid=param_grid,   # or param_grid_large
                  scoring='accuracy',
                  cv=cv,
                  n_jobs=-1,
                  verbose=2)
gs.fit(X_train, y_train)

print(gs.best_params_, gs.best_score_)

```

```python
rf = RandomForestClassifier(bootstrap= True, max_depth= None, max_features= 'sqrt', min_samples_leaf= 2, min_samples_split= 2, n_estimators= 100)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
